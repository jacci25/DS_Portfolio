## Map data wrangling

First, the TA shapefile and the normalised data can be loaded as follows.

```r
library(sf)
 
geo <- st_read("data/territorial-authority-2023-generalised.shp")
data <- read.csv("data/Final_norming_data.csv")
```

```r
str(geo)
```

```r
str(data)
```

Both data can now be joined on TA (left join on geo to keep all the geo data for each TA).

```r
# merge on TA
df <- merge(geo, data, by.x="TA2023_V_2", by.y="Territorial_Authority", all.x=TRUE)
```

```r
str(df)
```

The resulting dataframe can now be saved as a shapefile again.

```r
# Save as shapefile again
st_write(df, "data/Final_map.shp")
```
