```r
# load required libraries
library(dplyr)
library("readxl")
```

```r
# read Excel file
crime_data <- read_excel("Data/crime data.xlsx")
crime_data
```

```r
# print the content of the crime_data data frame and its structure (data types of columns)
str(crime_data)
```

```r
# select specific columns for further analysis
selected_crime_data <- crime_data[c("Territorial Authority", "Anzsoc Division", "Year Month", "Victimisations")]  
selected_crime_data
```

```r
# Rename the column to remove spaces
names(crime_data)[names(crime_data) == "Territorial Authority"] <- "Territorial_Authority"
names(crime_data)[names(crime_data) == "Anzsoc Division"] <- "Anzsoc_Division"

# Remove the trailing period ('.') from the "Territorial_Authority" column
crime_data$Territorial_Authority <- gsub("\\.$", "", crime_data$Territorial_Authority)

# Print the updated unique values
unique_territorial_authorities <- unique(crime_data$Territorial_Authority)

unique_territorial_authorities
```

```r
# group the data by "Territorial_Authority" and "Anzsoc_Division" 
# then calculates the total victimizations for each group
victims_by_TA <- crime_data %>%
  group_by(Territorial_Authority, Anzsoc_Division) %>%
  summarise(Total_Victimisations = sum(Victimisations))

victims_by_TA
```

```r
# note that there is a wrong spelling of Whanganui District
# change the wrong spelling to the correct

victims_by_TA$Territorial_Authority <- gsub("Wanganui District", "Whanganui District", victims_by_TA$Territorial_Authority)

# Checking all unique value and count in the house data
unique_values_TA <- unique(victims_by_TA$Territorial_Authority)
unique_values_TA <- sort(unique_values_TA)
unique_values_TA
length(unique_values_TA)
```

```r
# write the aggregated data in the victims_by_TA data frame to a CSV file 
victims_by_TA %>% write.csv("Data/Final_crime_by_TA.csv")
```
