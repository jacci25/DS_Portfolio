```r
library(tidyverse)
```

```r
row_school_data <- read.csv("Data/School_Data.csv")
# read the csv file
# I've hidden the first 15 columns of data because they are header introduction in the original CSV data. 
# Doing this allows me to obtain the DataFrame more quickly and ensures that each column has the correct data type.
```

```r
row_school_data
# check the dataframe 
```

```r
str(row_school_data)
# Check the data type of each column to ensure that there are no errors when using the "groupby" function later.
```

```r
school_data <- row_school_data[c("Territorial.Authority","School.Type","Total.School.Roll")]
school_data
# select the three columns of Territorial.Authority, School.Type, Total.School.Roll from the row_school_data
```

```r
# Remove the portion after " - " in the 'Territorial Authority' column of 'school_data'
# This step is to standardize the value in the 'TA' column in order to join with other tables.

school_data$Territorial.Authority <- gsub(" - .*", "", school_data$Territorial.Authority)
school_data
```

```r
# Get the unique values in the "Territorial.Authority" column of the "school_data" dataset
unique_values_TA <- unique(school_data$Territorial.Authority)

# Display the unique values
unique_values_TA

# Calculate and display the number of unique values
length(unique_values_TA)

# this chunk, I'd like to know how many Territorial Authorities in the dataframe,and the unique data
```

```r
# Get the unique values in the "School.Type" column of the "school_data" dataset
unique_values_school_type <- unique(school_data$School.Type)

# Display the unique values
unique_values_school_type

# Calculate and display the number of unique values
length(unique_values_school_type)

# this chunk, I'd like to know how many school types in the dataframe,and the unique data
```

```r
# Grouping the school_data by Territorial Authority and School Type
group_by_data_draft <- school_data %>%
  group_by(Territorial.Authority, School.Type) %>%
  
# Summarizing the grouped data by counting the number of schools and calculating the total school roll
  summarize(School_Count = n(), Total_School_Roll = sum(Total.School.Roll))

# Displaying the resulting summarized data
group_by_data_draft
```

```r
# Removing the first 5 rows from the data frame.
# These rows contain missing TA values and school roll counts of 0.

# Additionally, the correspondence school, which serves as the headquarters for several other schools,
# doesn't have a specific TA. Hence, it's excluded from the data frame

group_by_data <- group_by_data_draft[-c(1:5), ]
group_by_data
```

```r
# Replace the column name to standardise with others tables for easy joining

school_count <- select(group_by_data, 
                       Territorial_Authority = Territorial.Authority,
                       School_Type = School.Type, 
                       School_Count) 
    
school_count
```

```r
# Save the 'school_count' dataframe to a CSV file
school_count %>%
    write_csv("Data/Final_School_Data.csv")
```
