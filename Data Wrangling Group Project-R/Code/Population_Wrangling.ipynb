## Population data wrangling

Loading the excel file and specifying which tab to use.

```r
library("readxl")
```

```r
pop_data <- read_excel("data/Subnational-population-estimates-At-30-June-2022-provisional.xlsx", sheet = "Table 4")
pop_data
```

Removing the 'header' and 'footer' rows.

```r
library("dplyr")
pop_data_clean <- slice(pop_data, 4:(n() - 11))     
pop_data_clean
```

Renaming the columns and removing the rows containing the old column names.

```r
names(pop_data_clean) <- c("Territorial_Authority", "Auckland_Local_Board", "Year", "Population_Group_0-14", 
                           "Population_Group_15-39", "Population_Group_40-64", "Population_Group_65+", 
                           "Total_Population", "Population_Group_0-14_Percent", 
                           "Population_Group_15-39_Percent", "Population_Group_40-64_Percent", "Population_Group_65+_Percent", 
                           "Median_Age_Years")
pop_data_clean <- pop_data_clean[-(1:3), ]
pop_data_clean
```

As we are only interested in the territorial authorities, the rows dedicated to detailing the Auckland population per local board can be removed. 

```r
pop_data_clean <- pop_data_clean %>% filter(!is.na(Territorial_Authority))
pop_data_clean
```

The year values sometimes contain a "P". This can be removed as follows.

```r
pop_data_clean$Year <- gsub('P','',pop_data_clean$Year)
pop_data_clean$Year <- as.numeric(pop_data_clean$Year)
pop_data_clean
```

The data can now be filtered to only contain 2022 values, and only the relevant columns and rows can be selected.

```r
pop_data_final <- pop_data_clean %>%
                  filter(Year == 2022)
pop_data_final
```

```r
pop_data_final <- pop_data_final %>%
    select(Territorial_Authority, Total_Population)
pop_data_final
```

The last row containing a summary of New Zealand can also be removed.

```r
pop_data_final <- slice(pop_data_final, 1:(n() - 1)) 
```

The Territorial Authority column also contains some special characters that should be removed for consistency and easier joining.

```r
pop_data_final$Territorial_Authority <- gsub("Ō", "o", pop_data_final$Territorial_Authority)
pop_data_final$Territorial_Authority<- gsub("ō", "o", pop_data_final$Territorial_Authority)
pop_data_final$Territorial_Authority <- gsub("ā", "a", pop_data_final$Territorial_Authority)
pop_data_final$Territorial_Authority <- gsub("\\b(\\w)", "\\U\\1", tolower(pop_data_final$Territorial_Authority), perl = TRUE)
pop_data_final$Territorial_Authority <- gsub("Central Hawke\'S Bay District", "Central Hawke\'s Bay District", pop_data_final$Territorial_Authority)
pop_data_final$Territorial_Authority <- gsub("Western Bay Of Plenty District", "Western Bay of Plenty District", pop_data_final$Territorial_Authority)

pop_data_final
```

```r
Finally, the unique values of the Territorial Authority column can be checked.
```

```r
# Checking all unique value and count in the population data
unique_values_TA <- unique(pop_data_final$Territorial_Authority)
unique_values_TA <- sort(unique_values_TA)
unique_values_TA
length(unique_values_TA)
```

The data can be written as a csv as follows.

```r
library(tidyverse)
pop_data_final %>%
    write_csv("Data/Final_Population_Data.csv")
```
